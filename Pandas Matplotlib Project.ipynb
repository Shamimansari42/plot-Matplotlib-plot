```python
#An application that creates a number of classes and compares them using line chat and scatter plots

#module
from matplotlib import pyplot as plt
import numpy as np

# It increases the chat size
plt.figure(figsize=(9,9))

#number of classes 
try:
    no_of_classes = int(input('Number of Classes : '))
except ValueError:      # This error occurs when you try to store Alphabate
    raise print('Only numbers are allowed')

#Class percentages are stored in dictionary format using clsses_dict
classes_dict={}
for i in range(1,no_of_classes + 1):
    classes_dict[f'class_{i}']= ''

#A loop is used to collect input from the user: percentages
i = 1
while i <= len(classes_dict):

    #User input is taken
    user = list(input(f"Enter a class {i} percentage (seprated by ',') : ").split(','))
    if len(user) > len(classes_dict) or len(user) < len(classes_dict):                       # It will work if the number of classes is not equal to the number of percentages
        print('Number of classes must be equal to number of percentage')
    elif type(user) == str: # If you are storing alphabate in the list, it will work
        print('You can not enter Alphabates')
    else:
        user_list = [int(i) for i in user] # All string elements will be converted to integers
        classes_dict[f'class_{i}']= user_list
        i +=1

#colour of bar chat
colour = ['r','g','k','b','p', 'y', 'm']

#index or x axis elements for chat
classes_index = [f'Student {i}' for i in range(1, no_of_classes + 1)]
print(classes_index)

#dividing chat into subchat
plt.subplot(2,1,1)
plt.title('Class comparison')
plt.xlabel('Percentage')
plt.ylabel('Number of students')
for i in range(1, len(classes_dict)+1): # loop to create line chat
    plt.plot(classes_dict[f'class_{i}'], classes_index, color=colour[i-1], label=f'class {i}')
plt.legend(loc=1)

plt.subplot(2,1,2)
plt.title('Class comparison')
plt.xlabel('Percentage')
plt.ylabel('Number of students')
for i in range(1, len(classes_dict)+1): # loop to create scatter chat
    plt.scatter(classes_dict[f'class_{i}'], classes_index, label=f'class {i}')
plt.legend(loc=1)
plt.show()

```
